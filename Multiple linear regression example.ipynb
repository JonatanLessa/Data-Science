```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
base = pd.read_csv('Real_estate.csv')
base.shape
```

```python
base.head()
```

```python
base = base.drop(['No'], axis = 1)
base.head()
```

```python
base.rename(columns={'X1 transaction date': 'transaction_date', 'X2 house age': 'house_age', 'X3 distance to the nearest MRT station': 'dist_MRT', 'X4 number of convenience stores': 'number_of_convenience_stores', 'X5 latitude': 'latitude', 'X6 longitude': 'longitude', 'Y house price of unit area': 'price'}, inplace = True)
base.head()
```

```python
X = base.iloc[:, 1:6].values
y = base.iloc[:, 6].values
#X
y
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
#score R^2
model.score(X, y)
```

```python
new_model = np.array([13.3, 561.98450, 5, 24.98746, 121.54391])
new_model = new_model.reshape(1, -1)
model.predict(new_model)
```

```python
#Creating the model using the statsmodel library
fit_model = sm.ols(formula = 'price ~ house_age + dist_MRT + number_of_convenience_stores + X5_latitude + longitude', data = base)
trained_model = fit_model.fit()
trained_model.summary()
```
